```python
import pandas as pd
import numpy as np

# 0. Data and pre-processing
df = pd.read_csv('vitalsign.csv.gz')
print("Original Data")
print(df.head())

df = df.drop(columns=['charttime', 'o2sat', 'rhythm', 'pain'])
df = df.dropna()

print("\nAfter Pre-processing")
print(df.head())

# 1. Roll-up: by subject and stay, average all vitals
print("\n1. Roll-up: by subject and stay, average all vitals")
df_rollup = df.groupby(['subject_id', 'stay_id']).agg(
    average_temperature=('temperature', 'mean'),
    average_heartrate=('heartrate', 'mean'),
    average_resprate=('resprate', 'mean'),
    average_sbp=('sbp', 'mean'),
    average_dbp=('dbp', 'mean')
).reset_index()
print(df_rollup)

# 2. Slicing: by stay 34391979
print("\n2. Slicing: by stay 34391979")
df_slice = df[df['stay_id'] == 34391979]
print(df_slice)

# 3. Dicing: by subject 10014354 and 10002930, SBP >= 140 and DBP >= 90
print("\n3. Dicing: by subject 10014354 and 10002930, SBP >= 140 and DBP >= 90")
df_dice = df[(df['subject_id'].isin([10014354, 10002930])) & (df['sbp'] >= 140) & (df['dbp'] >= 90)]
print(df_dice)
```
